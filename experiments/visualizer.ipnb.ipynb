```python
from game.json_utils import load_numba_dict
from game.usage_utils import table_to_bin
```

```python
lookup_table = load_numba_dict("../data/lookup_table.json")
```

```python
def print_board(x_board, o_board):
    for i in range(9):
        if (x_board >> i) & 1:
            c = "X"
        elif (o_board >> i) & 1:
            c = "O"
        else:
            c = "_"
        print(c, end=" ")
        if i % 3 == 2:
            print()
```

```python
curr_board = table_to_bin([
    [" ", " ", " "],
    [" ", " ", " "],
    [" ", " ", " "]
])
```

```python
print(curr_board)
print_board(*curr_board)
ans = lookup_table[curr_board][1].bit_length() - 1
print(f"({ans % 3}, {ans // 3})")
```
